```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#Se importa las librerias de DBSCAN
#Sklearn.cluster nos ayudara a la creacion de los clusters
from sklearn.cluster import DBSCAN

#Se cargan los datos
df = pd.read_csv("/workspaces/repos1/Semana_11/Mall_Customers.csv")

df.head(10)
```

```python
#Se agregan a x las columnas que se desean trabajar
x = df.loc[:, ['Annual Income (k$)', 'Spending Score (1-100)']].values
```

```python
#Se crea un "vector" para guardar el resultado del clusterizado
#Se define el epsilon en el parametro eps, y el minimo de puntos minimos para saber si hay densidad en parametro min_samples
clusters = DBSCAN(eps = 8, min_samples = 4).fit(x)
labels = clusters.labels_

#Se muestra el resultado clusters
print(labels)
```

```python
# Gráfica de matplotlib para mostrar Los CLusters
plt.scatter(x[:, 0], x[:,1], c = labels, cmap= "plasma")
plt.xlabel("Income")
plt.ylabel("Spending Score")
plt.show()
```
