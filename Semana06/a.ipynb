```python
# Separamos las columnas creando treatment para ubicar el tipo de dosis y columna dose para ubicar cada dosis 
# Quitamos las dosis que tienen solo un “-” como valor  
df = pd.read_csv('/workspaces/repos/Semana06/treatments.csv')
df = df.dropna()
df = pd.melt(df, id_vars=['given_name', 'surname', 'hba1c_start', 'hba1c_end', 'hba1c_change'], var_name='treatment', value_name='dose')
df = df[df.dose != "-"]
df
```

```python
# Separamos las dosis para ubicar el inicio y final de estas, y usamos drop para eliminar la columna dose
df[['dose_start', 'dose_end']] = df['dose'].str.extract(r'(.+) - (.+)')
df = df.drop('dose', axis=1)
df
```
