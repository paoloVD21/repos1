```python
import pandas as pd
import numpy as np
from scipy import stats

df = pd.read_csv('/workspaces/repos/Semana_05/data_base_1.csv')
df_new = df[(np.abs(stats.zscore(df.SUCRE)) < 3)]
df_new

q_low = df["SUCRE"].quantile(0.01)
q_hi  = df["SUCRE"].quantile(0.99)
 
#filtering outliers
df_filtered = df[(df["SUCRE"] < q_hi) & (df["SUCRE"] > q_low)]
df_filtered
```
