```python
#TRANSFORMACIÃ“N DE DATOS
#Cambiamos algunos nombres de los indices de ingles a espaÃ±ol
df.columns = ['Genero', 'Edad', 'Altura', 'Peso', 'Historia_familiar_con_sobrepeso', 'FAVC', 'FCVC', 'NCP', 'CAEC', 'Fuman', 'CH2O', 'SCC', 'FAF', 'TUE', 'CALC', 'Medio_de_transporte', 'NObeyesdad']
df
```

```python
#Cambiamos palabras de ingles a espaÃ±ol en el data set
df.loc[df['Genero'] == 'Female', 'Genero'] = 'Femenino'
df.loc[df['Genero'] == 'Male', 'Genero'] = 'Masculino'
df.loc[df['Historia_familiar_con_sobrepeso'] == 'yes', 'Historia_familiar_con_sobrepeso'] = 'si'
df.loc[df['FAVC'] == 'yes', 'FAVC'] = 'si'
df.loc[df['CAEC'] == 'Sometimes', 'CAEC'] = 'A veces'
df.loc[df['CALC'] == 'Always', 'CALC'] = 'Siempre'
df.loc[df['CAEC'] == 'Frequently', 'CAEC'] = 'Frecuentemente'
df.loc[df['Fuman'] == 'yes', 'Fuman'] = 'si'
df.loc[df['SCC'] == 'yes', 'SCC'] = 'si'
df.loc[df['CALC'] == 'Sometimes', 'CALC'] = 'A veces'
df.loc[df['CALC'] == 'Frequently', 'CALC'] = 'Frecuentemente'
df.loc[df['Medio_de_transporte'] == 'Public_Transportation', 'Medio_de_transporte'] = 'Trasporte_publico'
df.loc[df['Medio_de_transporte'] == 'Walking', 'Medio_de_transporte'] = 'Caminando'
df.loc[df['Medio_de_transporte'] == 'Automobile', 'Medio_de_transporte'] = 'Automovil'
df.loc[df['Medio_de_transporte'] == 'Bike', 'Medio_de_transporte'] = 'Bicicleta'
df.loc[df['NObeyesdad'] == 'Normal_Weight', 'NObeyesdad'] = 'Peso_normal'
df.loc[df['NObeyesdad'] == 'Insufficient_Weight', 'NObeyesdad'] = 'Peso_insufuciente'
df.loc[df['NObeyesdad'] == 'Obesity_Type_I', 'NObeyesdad'] = 'Obesidad_tipo_I'
df.loc[df['NObeyesdad'] == 'Obesity_Type_II', 'NObeyesdad'] = 'Obesidad_tipo_II'
df.loc[df['NObeyesdad'] == 'Obesity_Type_III', 'NObeyesdad'] = 'Obesidad_tipo_III'
df.loc[df['NObeyesdad'] == 'Overweight_Level_I', 'NObeyesdad'] = 'Nivel_de_sobrepeso_I'
df.loc[df['NObeyesdad'] == 'Overweight_Level_II', 'NObeyesdad'] = 'Nivel_de_sobrepeso_II'
df.loc[df['NObeyesdad'] == 'Overweight_Level_III', 'NObeyesdad'] = 'Nivel_de_sobrepeso_III'

df
```

```python
#Cambiar tipo de dato de una columna
df['Edad'] = df['Edad'].apply(int)
df['Altura'] = df['Altura'].astype(float).round(2)
df['Peso'] = df['Peso'].astype(float).round(2)
df['CH2O'] = df['CH2O'].apply(int)
df['FAF'] = df['FAF'].apply(int)
df['TUE'] = df['TUE'].apply(int)

df
```

```python
# COMANDO PARA MOSTRAR EL DIAGRAMA DE BIGOTES
plt.boxplot(df['Peso'], vert=False)   
```

```python
#COMANDO PARA MOSTRAR DIAGRAMA DE BARRAS DE PROMEDIO DE PESOS DE HOMBRES Y MUJERES
promedios = df.groupby('Genero')['Peso'].mean()

# Crea el grÃ¡fico de barras
promedios.plot(kind='bar', color=['blue', 'pink'])
plt.title('Promedio de Pesos por GÃ©nero')
plt.xlabel('GÃ©nero')
plt.ylabel('Promedio de Peso')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.scatter(df['Edad'], df['Peso'], alpha=0.5)
plt.title('RelaciÃ³n entre Edad y Peso')
plt.xlabel('Edad')
plt.ylabel('Peso')
plt.grid(True)
plt.show()
```
