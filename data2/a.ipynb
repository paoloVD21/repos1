```python
import pandas as pd
import glob
import random
import os
all_files = glob.glob('/workspaces/repos/data2/Aemet2019-*.xls')

file_list = []

for f in all_files:
    data = pd.read_excel(f)
    data['source_file'] = f
    file_list.append(data)

df = pd.concat(file_list, ignore_index = True)

df
```

```python
filas , columnas = df.shape
print("Numero de filas: ", filas)
print("Numero de columnas: ", columnas)
```

```python
primeras_10_filas = df.head(10)
print(primeras_10_filas)
```

```python
nombre_columnas = {
    'Estación': 'estacion','Provincia':'provincia','Temperatura máxima (ºC)':'temp_max','Temperatura mínima (ºC)':'temp_min','Temperatura media (ºC)':'temp_med','Racha (km/h)':'viento_racha','Velocidad máxima (km/h)':'viento_vel_max','Precipitación 00-24h (mm)':'prec_dia','Precipitación 00-06h (mm)':'prec_0_6h','Precipitación 06-12h (mm)':'prec_6_12h','Precipitación 12-18h (mm)':'prec_12_18h','Precipitación 18-24h (mm)':'prec_18_24h','source_file':'fecha'}
df.rename(columns=nombre_columnas, inplace=True)
df
```
