```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import seaborn as sns
import csv
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
from bs4 import BeautifulSoup
import requests
```

```python
# URL y Headers para saltar seguridad 
website = "https://datosmacro.expansion.com/demografia/poblacion/usa"
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}

resultado = requests.get(website, headers=headers)
soup = BeautifulSoup(resultado.text, 'html.parser')
print(soup.prettify())

# Encontrar todas las filas dentro de la tabla
data_rows = soup.select('table tr')  # Esto selecciona todas las filas dentro de la tabla

# Lista para almacenar los datos
data_list = []

# Encabezados de la tabla
headers = ['Fecha', 'Densidad', 'Hombres','Mujeres', 'Población']
data_list.append(headers)

# Iterar sobre las filas y extraer los datos
for row in data_rows:
    # Extraer datos de cada celda en la fila
    cells = row.find_all('td')
    
    # Obtener texto de las celdas y limpiarlo
    data = [cell.get_text(strip=True) for cell in cells]
    
    # Agregar datos a la lista
    data_list.append(data)

# Escribir datos en un archivo CSV
with open('EstadosUnidosPoblacion.csv', 'w', newline='', encoding='utf-8') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerows(data_list)
```

```python
df_Peru['id_Pais'] = 3
df_Peru
```

```python
# Integrar los tres DataFrames
dfs = [df_EeUu, df_Brasil, df_Peru]
df_final = pd.concat(dfs, ignore_index=True)
df_final
```

```python
#Gráfico de Línea (Line Plot) para Población a lo largo del tiempo
plt.figure(figsize=(12, 6))
for pais in df_final['id_Pais'].unique():
    df_pais = df_final[df_final['id_Pais'] == pais]
    plt.plot(df_pais['Año'], df_pais['Poblacion'], label=f'País {pais}')
plt.gca().yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: '{:.0f}M'.format(x / 1e6)))
plt.title('Población a lo largo del tiempo')
plt.xlabel('Año')
plt.ylabel('Población (en millones)')
plt.legend()
plt.show()
```

```python
#Gráfico de Violín (Violin Plot) para Población por País:

plt.figure(figsize=(12, 6))
ax = sns.violinplot(x='id_Pais', y='Poblacion', data=df_final)
ax.yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: '{:.0f}M'.format(x / 1e6)))
plt.title('Distribución de Población por País')
plt.xlabel('ID de País')
plt.ylabel('Población (en millones)')
plt.show()
```

```python
#Gráfico de Dispersión (Scatter Plot) para Año vs. Población:
plt.figure(figsize=(12, 6))
ax = sns.scatterplot(x='Año', y='Poblacion', hue='id_Pais', data=df_final)
# Establece el formato del eje y en notación de millones
ax.yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: '{:.0f}M'.format(x / 1e6)))
plt.title('Año vs. Población por País')
plt.xlabel('Año')
plt.ylabel('Población (en millones)')
plt.legend(title='ID de País')
plt.show()
```

```python
#Gráfico de Barras (Bar Plot) para Población por País:
ax = sns.barplot(x='id_Pais', y='Poblacion', data=df_final)

# Establece el formato del eje y en notación de millones
ax.yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: '{:.0f}M'.format(x / 1e6)))

plt.title('Población por País')
plt.xlabel('ID de País')
plt.ylabel('Población (en millones)')
plt.show()
```

```python
#Gráfico Density Plot para Población por País:
plt.figure(figsize=(12, 6))
sns.kdeplot(data=df_final, x='Año', y='Poblacion', hue='id_Pais', fill=True, common_norm=False)
plt.gca().yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: '{:.0f}M'.format(x / 1e6)))
plt.title('Density Plot de Población a lo largo del tiempo por País')
plt.xlabel('Año')
plt.ylabel('Población (en millones)')
plt.legend(title='ID de País')
plt.show()
```
